```python
containers = [x.get_attribute("innerHTML") for x in browser.find_elements(By.XPATH,"//div[@class='review-container']")]
```

```python
len(containers)
```

```python
containers[1]
```

```python
from bs4 import BeautifulSoup
import re
```

```python
reviews = []
ratings = []

for container in containers:
  bsObj = BeautifulSoup(container)
  rev = bsObj.find(name = "div", attrs = {"class":"text show-more__control"}).text
  reviews.append(rev)

  try:
    rat = bsObj.find(name = "span", attrs = {"class":"rating-other-user-rating"}).text
    rat = re.sub("\n","",rat)
    rat = re.sub("/10","",rat)
    ratings.append(rat)
  except:
    ratings.append(np.nan)
```

```python
df = pd.DataFrame(columns = ["review","rating"])
df["review"] = reviews
df["rating"] = ratings
df.head()
```
